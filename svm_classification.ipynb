```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import cross_validate
from sklearn.metrics import balanced_accuracy_score
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=52)
s2 = StratifiedShuffleSplit(n_splits=5, test_size=0.1, random_state=52)
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
for i, (train_index, test_index) in enumerate(sss.split(X, y)):
    X_train = dataset.iloc[train_index,:-1]
    y_train = dataset.iloc[train_index,-1]
    for i, (train_index_c, test_index_c) in enumerate(s2.split(X_train, y_train)):
        X_train_c = X_train.iloc[train_index_c,:]
        y_train_c = y_train.iloc[train_index_c]
        #print(y_train_c)
        clf.fit(X_train_c,y_train_c)

        X_test_c = X_train.iloc[test_index_c]
        y_test_c = y_train.iloc[test_index_c]
        print(clf.score(X_train_c,y_train_c),clf.score(X_test_c,y_test_c))
    # X_test = dataset.iloc[test_index,:-1]
    # y_test = dataset.iloc[test_index,-1]
    # print(clf.score(X_train,y_train),clf.score(X_test,y_test))
```
